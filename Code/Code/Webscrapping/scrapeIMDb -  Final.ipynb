```python
def getPageN(url):

    page_infs = []

    try:        
        source = requests.get(url)

        source.raise_for_status()

        soup = BeautifulSoup(source.text, "html.parser")
        
        movies = soup.find('div', class_ = "lister-list").find_all("div",class_="lister-item mode-advanced")

        for movie in movies:

            infs = {}

            try:

                lic = movie.find("div", class_="lister-item-content")

                try:
                    number = lic.find("span", class_="lister-item-index unbold text-primary").text
                except Exception:
                    number = "NaN"
                infs["number"] = number

                try:
                    name = lic.a.text
                except Exception:
                    name = "NaN"
                infs["name"] = name

                try:
                    year = lic.find("span", class_="lister-item-year text-muted unbold").text
                except Exception:
                    year = "NaN"
                infs["year"] = year

                try:
                    certificate = lic.find("span", class_="certificate").text
                except Exception:
                    certificate = "NaN"
                infs["certificate"] = certificate

                try:
                    runtime = lic.find("span", class_="runtime").text
                except Exception as e:
                    runtime = "NaN"
                infs["runtime"] = runtime

                try:
                    genre = lic.find("span", class_="genre").text
                except Exception:
                    genre = "NaN"
                infs["genre"] = genre

                try:
                    rating = lic.find("div", class_="inline-block ratings-imdb-rating").strong.text
                except Exception:
                    rating = "NaN"
                infs["rating"] = rating

                try:
                    metascore = lic.find("div", class_="inline-block ratings-metascore").text
                except Exception:
                    metascore = "NaN"
                infs["metascore"] = metascore

                try:
                    text_muted = lic.find_all("p", class_="text-muted")[1].text
                except Exception:
                    text_muted = "NaN"
                infs["text-muted"] = text_muted
                
                try:
                    l = np.array(lic.find_all("p")[2].text.split("\n"))
                    i = argmax(l==['    Director:']*len(l))
                    j = argmax(l==['    Stars:']*len(l))
                    director = l[i+1:j]
                    stars = l[j+1:]

                except Exception as e:
                    # print(e)
                    director = "NaN"
                    stars = "NaN"

                infs["director"] = director
                infs["stars"] = stars
                                

            except Exception as e:
                print(e, "loop")
            
            # print(infs)
            page_infs += [infs]

    except Exception as e:
        print(e,"whole")

    return page_infs
```

```python
def cleanPage(films):
    
    films.drop('number', axis = 1, inplace=True)
    
    films['year'] = films["year"].apply(lambda x : x[1:-1] )
    
    films['runtime'] = films['runtime'].apply(lambda x : int(x.split(" ")[0]))
    films = films.astype({'runtime':'int'})
    
    films['genre'] = films['genre'].apply(lambda x : x[1:])
    films['genre'] = films['genre'].apply(lambda x : [g for g in x.split(',')])

    films = films.astype({'rating':'float'})

    films['metascore'] = films['metascore'].apply(lambda x : int(re.search(r'\d+', x).group()) if re.search(r'\d+', x) != None else -1)

    films["text-muted"] = films["text-muted"].apply(lambda x: x[1:])

    films["director"] = films["director"].apply(lambda x : [ d if d != '| ' else None for d in x] )

    films["stars"] = films["stars"].apply(lambda x : [s.split(',') for s in x])
    films["stars"] = films["stars"].apply(lambda x : [s[0] for s in x[:-1]])

    return films
```

```python
def get_genre(genre):
    #collect over all pages
    df = []
    n_start = 1
    step = 50

    # get the number of films
    url = f"https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&genres={genre}&languages=en&sort=user_rating,desc&start=1&explore=genres&ref_=adv_nxt"
    source = requests.get(url)
    source.raise_for_status()
    soup = BeautifulSoup(source.text, "html.parser")
     
    num=soup.find('div',class_="desc")
    nb_Films = 0
    try :
        line=num.find("span").text
        nb_Films= int(line.split()[2].replace(',',''))
    except Exception:
        print("Number of films not found")
    
    print("Number of films =", nb_Films)
    print(type(nb_Films))
    
    ite = int(np.ceil((nb_Films-n_start)/step))
    
    for i in range(ite):
        url = f"https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&genres={genre}&languages=en&sort=user_rating,desc&start={ite*i+n_start}&explore=genres&ref_=adv_nxt"
        page = getPageN(url)
        
        df += [page]

        print(n_start+i*step)

    t_df = pd.concat([pd.DataFrame(d) for d in df])
    t_df = cleanPage(t_df)
    t_df.to_csv(f"./data_imdb_{genre}.csv", sep=";")
```
