```python
directors = []
for i in range(df.shape[0]):
    st = df['director'].loc[i]
    st = st.replace("'    Directors:',", "")
    st = st.replace(", ", "")
    st = st.replace("None", "")
    st = st.replace("''", ",")
    st = st.replace('[',"")
    st = st.replace(']',"")
    st = st.replace("'", "")
    st = st.split(",")
    directors.append(st)
    
directors
```

```python
genres = []
for i in range(df.shape[0]):
    st = df['genre'].loc[i]
    st = st.replace(" ", "")
    st = st.replace('[',"")
    st = st.replace(']',"")
    st = st.replace("'", "")
    st = st.split(",")
    genres.append(st)

genres
```

```python
stars = []
for i in range(df.shape[0]):
    st = df['stars'].loc[i]
    st = st.replace("'    Director:',", "")
    st = st.replace('[',"")
    st = st.replace(']',"")
    st = st.replace('"', "")
    st = st.replace("'", "")
    st = st.split(", ")
    stars.append(st)
        
stars
```

```python
df['director'] = directors
df['genre'] = genres
df['stars'] = stars
df
```

```python
stars2 = []
for i in range(len(stars)):
    for j in range(len(stars[i])):
        stars2.append(stars[i][j])

stars2
```

```python
dfstars = pd.DataFrame(stars2, columns=['stars'])
dfstars["count"] = 1
```
